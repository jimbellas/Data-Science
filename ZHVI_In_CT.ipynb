```python

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt


#This CSV file comes from Zillow.com
#It contain data about their Zillow Home Value Index (ZHVI) which is "A measure of the typical home value and market changes across a given region and housing type"
df = pd.read_csv (r'C:\Users\Jim\iCloudDrive\Documents\Code\Zillow_rating.csv')

#This creates a data frame with only datapoints from CT
CTdf = df.query('StateName == "CT"')
##print(CTdf.describe)
#From this we can see a summary of the data incuding the mean ZHVI for the state of CT

RegionName_df = CTdf[['RegionName']]
Y2023CT_df = CTdf[['2023-11-30','2023-10-31','2023-11-30','2023-09-30','2023-08-31','2023-07-31','2023-06-30','2023-05-31','2023-04-30','2023-03-31','2023-02-28','2023-01-31']]
Y2022CT_df = CTdf[['2022-12-31','2022-11-30','2022-10-31','2022-11-30','2022-09-30','2022-08-31','2022-07-31','2022-06-30','2022-05-31','2022-04-30','2022-03-31','2022-02-28','2022-01-31']]
Y2021CT_df = CTdf[['2021-12-31','2021-11-30','2021-10-31','2021-11-30','2021-09-30','2021-08-31','2021-07-31','2021-06-30','2021-05-31','2021-04-30','2021-03-31','2021-02-28','2021-01-31']]
Y2020CT_df = CTdf[['2020-12-31','2020-11-30','2020-10-31','2020-11-30','2020-09-30','2020-08-31','2020-07-31','2020-06-30','2020-05-31','2020-04-30','2020-03-31','2020-02-29','2020-01-31']]

df4 = pd.concat([RegionName_df,Y2023CT_df,Y2022CT_df,Y2021CT_df,Y2020CT_df], axis = 1)

df_melted = pd.melt(df4, id_vars=['RegionName'], var_name='Date', value_name='Value')
df_melted['Date'] = pd.to_datetime(df_melted['Date'])
plt.figure(figsize=(20, 12))

df5=df4.T

for region in df_melted['RegionName'].unique():
    region_data = df_melted[df_melted['RegionName'] == region]
    plt.plot(region_data['Date'], region_data['Value'], label=region, marker='o')

plt.xlabel('Date')
plt.ylabel('ZHVI Price')
plt.title('ZHVI Over Time by Region')
plt.legend()
plt.grid(True)




plt.show()



```
